```python
# ================================================
# MONTE CARLO OPTION PRICING (GOOGLE COLAB READY)
# ================================================

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def monte_carlo_option_pricing(S0, K, r, sigma, T, N):
    """
    Monte Carlo simulation for European option pricing using Geometric Brownian Motion
    """

    # Generate random Z ~ N(0,1)
    Z = np.random.standard_normal(N)

    # Simulate stock prices using GBM formula
    stock_prices = S0 * np.exp((r - 0.5 * sigma**2) * T + sigma * np.sqrt(T) * Z)

    # Calculate expected stock price and standard deviation
    expected_price = np.mean(stock_prices)
    std_price = np.std(stock_prices)

    # Option payoffs
    call_payoffs = np.maximum(stock_prices - K, 0)
    put_payoffs = np.maximum(K - stock_prices, 0)

    # Discounted prices
    call_price = np.exp(-r * T) * np.mean(call_payoffs)
    put_price = np.exp(-r * T) * np.mean(put_payoffs)

    # ================================================
    # HISTOGRAM (Purple)
    # ================================================
    plt.figure(figsize=(12, 8))

    plt.hist(
        stock_prices,
        bins=50,
        alpha=0.7,
        color='purple',      # <- new color
        edgecolor='black',
        linewidth=0.5,
        density=True,
        label='Simulated Stock Prices'
    )

    # Mean line
    plt.axvline(
        expected_price,
        color='red',
        linestyle='--',
        linewidth=3,
        label=f'Mean Stock Price: ${expected_price:.2f}'
    )

    # Strike price
    plt.axvline(
        K,
        color='green',
        linestyle='-',
        linewidth=2,
        label=f'Strike Price: ${K}'
    )

    # Labels & title
    plt.xlabel('Stock Price at Maturity ($)', fontsize=12, fontweight='bold')
    plt.ylabel('Probability Density', fontsize=12, fontweight='bold')
    plt.title('Monte Carlo Simulation: Distribution of Stock Prices at Maturity',
              fontsize=14, fontweight='bold')

    # Grid + legend
    plt.grid(True, alpha=0.3)
    plt.legend(fontsize=11)

    plt.tight_layout()
    plt.show()

    return {
        'expected_price': expected_price,
        'std_price': std_price,
        'call_price': call_price,
        'put_price': put_price,
        'stock_prices': stock_prices
    }


# ================================================
# DEMONSTRATION RUN IN COLAB
# ================================================
print("Running Monte Carlo Simulation (Colab version)...")

test_S0 = 100
test_K = 105
test_r = 0.05
test_sigma = 0.20
test_T = 1
test_N = 10000

results = monte_carlo_option_pricing(
    test_S0, test_K, test_r, test_sigma, test_T, test_N
)

print("\n--- Results ---")
print(f"Expected stock price: ${results['expected_price']:.2f}")
print(f"Std Dev: {results['std_price']:.2f}")
print(f"Call Price: ${results['call_price']:.4f}")
print(f"Put Price: ${results['put_price']:.4f}")
print(f"\nFunction successful!")
```

```python
# =============================================================
# INITIAL SETUP FOR GOOGLE COLAB
# =============================================================

import numpy as np
import matplotlib.pyplot as plt

# Make plots look clean
plt.style.use('default')


# =============================================================
# FUNCTION REQUIRED FOR QUESTION 3
# =============================================================

def monte_carlo_option_pricing(S0, K, r, sigma, T, N,
                               plot_histogram=False, label="Simulation"):
    """
    Runs Monte Carlo option pricing simulation.
    Returns dictionary with stock prices, expected price, std,
    call price, put price.
    """

    # Generate random samples
    Z = np.random.normal(0, 1, N)

    # Simulate terminal stock prices
    ST = S0 * np.exp((r - 0.5 * sigma**2) * T + sigma * np.sqrt(T) * Z)

    # Payoffs
    call_payoff = np.maximum(ST - K, 0)
    put_payoff = np.maximum(K - ST, 0)

    # Discounting
    call_price = np.exp(-r * T) * np.mean(call_payoff)
    put_price = np.exp(-r * T) * np.mean(put_payoff)

    # Plot histogram only if needed
    if plot_histogram:
        plt.hist(ST, bins=50, alpha=0.6, density=True)
        plt.axvline(K, color='black', linewidth=2, label=f"Strike Price K={K}")
        plt.title(label)
        plt.xlabel("Stock Price at Maturity")
        plt.ylabel("Density")
        plt.legend()
        plt.show()

    return {
        "stock_prices": ST,
        "expected_price": np.mean(ST),
        "std_price": np.std(ST),
        "call_price": call_price,
        "put_price": put_price
    }


# =============================================================
# QUESTION 3: VOLATILITY COMPARISON
# =============================================================

print("\n" + "=" * 60)
print("QUESTION 3: VOLATILITY COMPARISON")
print("=" * 60)

# Base parameters
S0 = 100
K = 105
r = 0.05
T = 1
N = 10000

# Test volatilities
volatilities = [0.10, 0.20, 0.30]
results_dict = {}

plt.figure(figsize=(12, 8))

for sigma in volatilities:

    results = monte_carlo_option_pricing(
        S0, K, r, sigma, T, N,
        plot_histogram=False,
        label=f" = {sigma*100}%"
    )

    # Store results
    results_dict[sigma] = results

    # Plot all histograms
    plt.hist(
        results['stock_prices'],
        bins=50,
        alpha=0.5,
        density=True,
        label=f'Volatility = {sigma*100:.0f}%',
        edgecolor='black'
    )

# Add strike price line
plt.axvline(K, color='red', linestyle='--', linewidth=3,
            label=f'Strike Price K={K}')

plt.xlabel('Stock Price at Maturity', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.title('Monte Carlo Simulation: Stock Price Distribution\nfor Different Volatilities',
          fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.show()


# =============================================================
# PRINT RESULTS TABLE
# =============================================================

print("\nVolatility Comparison Results:")
print("    | Expected Price | Std Dev    | Call Price | Put Price")
print("-" * 65)

for sigma in volatilities:
    res = results_dict[sigma]
    print(f"{sigma*100:2.0f}% | ${res['expected_price']:12.2f} | "
          f"${res['std_price']:9.2f} | ${res['call_price']:10.4f} | "
          f"${res['put_price']:9.4f}")


# =============================================================
# DISCUSSION OF EFFECTS OF VOLATILITY
# =============================================================

print("\n" + "=" * 60)
print("DISCUSSION: EFFECTS OF VOLATILITY")
print("=" * 60)

discussion = """
 When volatility increases, the distribution of possible future stock
  prices spreads out  basically, the stock behaves more "wildly."

 This wider spread raises both call and put option prices:
    - Calls benefit because high stock prices become more likely.
    - Puts benefit because low stock prices become more likely.

 Higher volatility = more uncertainty = more valuable options.

 In short, volatility is the heartbeat of option pricing.
"""

print(discussion)
```

```python
print("=" * 70)
print("QUESTION 4: ROLE OF RANDOM VARIABLE Z ~ N(0,1)")
print("=" * 70)

explanation = """
ROLE OF RANDOM VARIABLE Z ~ N(0,1) IN MONTE CARLO SIMULATION:

FINANCIAL INTERPRETATION:
- Z represents the random shocks or unexpected price movements in the stock market
- Each random draw of Z corresponds to one possible future scenario or market outcome
- The magnitude and direction of Z determine how much the actual stock return deviates
  from the expected return (r - 0.5)T

MATHEMATICAL ROLE IN GBM:
- In the Geometric Brownian Motion formula:
  S_T = S  exp((r - 0.5)T + T  Z)
- Z adds stochastic (random) component to the deterministic growth
- The term T  Z scales the randomness based on volatility and time horizon
- This ensures stock prices follow a log-normal distribution (always positive)

WHY Z IS CRUCIAL FOR MODELING UNCERTAINTY:

Models real-life uncertainty:
Markets move unpredictably. Z acts like the mathematical stand-in for that
chaos. Each random Z creates a different possible future stock price path.

Creates a distribution of outcomes:
Without Z, every stock would end up at one exact price. With Z, we get a whole
spread of outcomes  which is exactly what we need when pricing options, since
their value comes from uncertainty.

Captures volatility:
Z has standard deviation 1, but we stretch it by T.
Bigger   bigger price swings.
This is how volatility directly shapes the final stock distribution.

Keeps risk-neutral pricing valid:
We use risk-free rate

r for the drift but still use real-world volatility  through Z.
That balance ensures no-arbitrage pricing while still allowing uncertainty.

Enables Monte Carlo estimation:
By generating thousands of Z values, were effectively simulating thousands of
possible futures. Averaging the discounted payoffs across all those futures
gives us the option price.

In a nutshell:
Z is the heartbeat of the randomness in Monte Carlo simulations. Without it,
everything would be deterministic and boring no uncertainty, no payoff
variability, and honestly no point in pricing options at all. Using a normal
distribution for Z keeps most outcomes near expected levels, but still allows
those rare but important big jumps and crashes, just like real markets.
"""

print(explanation)
```
